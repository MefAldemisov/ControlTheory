```python
import numpy as np
import scipy.linalg as lin
import scipy.signal as sig
import scipy.integrate as diff
import matplotlib.pyplot as plt
```

```python
b = 0.9
m = 1
g = 9.81
l = 1

A_odd = np.array([[-b/m, -g/l], [1, 0]])
A_even = np.array([[-b/m, g/l],[1, 0]])

B = np.array([1, 0])
```

```python
def function(x, t, u):
    n = [-b/m*x[0] - g/l*np.sin(x[1]), x[0]]
    return  n - u


def observer(x_hat, t, u, y):
    return np.dot(A_odd, x_hat) + np.dot(B, u.reshape(-1, 1)) + L*(y - C*x_hat)
```

```python
C = 1
P = np.array([[0.1, 0.5],[0.2, 0]])

L = 10

dt = 1/1000
T = 20
time = np.linspace(0, T, dt**(-1))
x_0 = np.array([0.01, 0.001])
y = [C*x_0]
x_hat_arr = [np.array([0.1, 0.1])]
for i in range(1, len(time)):
#     TODO: use odeint between two dots,
#     BUT u is fixed between two poins
#     P controller u = P
    local_time = np.linspace(time[i-1], time[i])
    
    u = np.dot(P, y[-1])
    
    result = diff.odeint(function, x_0, local_time, args=tuple([u]))
    y.append(C*result[-1])
    x_0 = result[-1]
    result_hat = diff.odeint(observer, x_hat_arr[-1], local_time, args=tuple([u, y[-1]]))
    x_hat_arr.append(result_hat[-1])
#     print(y)
```

```python
y = np.array(y)
x_hat_arr = np.array(x_hat_arr)
```

```python
plt.figure(figsize=(8, 8))
plt.title("Even pendulum system simulation")
plt.plot(time, y.T[0], label="init velocity")
plt.plot(time, y.T[1], label="init coordinate")

plt.plot(time, x_hat_arr.T[0], label="obs velocity")
plt.plot(time, x_hat_arr.T[1], label="obs coordinate")
plt.grid()
plt.legend()
```
