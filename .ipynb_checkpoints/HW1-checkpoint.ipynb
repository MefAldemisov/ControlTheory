# From linear ODE to State Space

Given an ODE:

$$m\ddot x+b_{1}\dot x + k x= F$$

find its state space representation:

$$\dot x = Ax + b$$

```python
# task 1
F = 20
m = 25
b = 8
k = 24

# task 2

F = 0
b = 0
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

n = 3                     # degree of the polynomial 
a = np.array([m, b, k])     # coefficients: [a0 a1 ... ak]
a_norm = a[1:] / m     # divide by "ak"
A = np.zeros((n-1, n-1))  # state matrix
A[0 , 0:] = -a_norm
A[1:, 0:(n-2)] = np.eye(n-2)

B = np.zeros(A.shape[0])
B[0] = F/m

print("Our matrix A:", "Our matrix B:")
print(A, B)
```

```python
def equationSimplie(x, t):
    res = np.dot(A, x) +  B
    return res
```

```python
time = np.linspace(0, 40, 1000)   
# x0 = np.random.rand(n-1)*2 -1  # initial states
x0 = [0, 0.1]
res = odeint(equationSimplie, x0, time).T
```

```python
fig = plt.figure(figsize=(8, 8))
plt.title("Dipendence of veocity and coordinate in mass-spring system from time")
plt.xlabel("time")
plt.plot(time, res[0], label="velocity")
plt.plot(time, res[1], label="coordinate")
plt.grid()
plt.legend(shadow=True)
```

## Theoretical solution:
$$A = WDW^{-1}$$
$$W^{-1}\dot{x} = DW^{-1}x + W^{-1}b$$
<center>let B = $W^{-1}b$<br>
let R = $W^{-1}x$</center>

$$\dot{R} = DR + B$$
$$x = WR$$
