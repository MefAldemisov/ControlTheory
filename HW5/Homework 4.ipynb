```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Universal constants
g = 9.81
```

# The original nonlinear dynamics in State-Space form
$$
\dot{z}=\left[\begin{matrix}\dot{x}\\\dot{\theta}\\\begin{matrix}\ddot{x}\\\ddot{\theta}\\\end{matrix}\\\end{matrix}\right]=\left[\begin{matrix}\dot{x}\\\dot{\theta}\\\begin{matrix}\frac{-\sin{\left(\theta\right)}{\dot{\theta}}^2+g\sin{\left(\theta\right)}\cos{\left(\theta\right)}}{6-{\cos{\left(\theta\right)}}^2}\\\frac{-\cos{\left(\theta\right)}\sin{\left(\theta\right)}{\dot{\theta}}^2+6g\sin{\left(\theta\right)}}{6-{\cos{\left(\theta\right)}}^2}\\\end{matrix}\\\end{matrix}\right]+\left[\begin{matrix}0\\0\\\begin{matrix}\frac{1}{6-{\cos{\left(\theta\right)}}^2}\\\frac{\cos{\left(\theta\right)}}{6-{\cos{\left(\theta\right)}}^2}\\\end{matrix}\\\end{matrix}\right]u
$$

```python
def z_dot(z, t, u):
    x = z[0]
    theta = z[1]
    x_dot = z[2]
    theta_dot = z[3]
    x_dot_dot = (-np.sin(theta)*theta_dot**2+g*np.sin(theta)*np.cos(theta))/(6-np.cos(theta)**2) + u/(6-np.cos(theta)**2)
    theta_dot_dot = (-np.cos(theta)*np.sin(theta)*theta_dot**2+6*g*np.sin(theta))/(6-np.cos(theta)**2) + u*np.cos(theta)/(6-np.cos(theta)**2)
    return [x_dot, theta_dot, x_dot_dot, theta_dot_dot]
```

```python
# Simulating it without any control
time = np.linspace(0, 10, 1000)
z_0 = [1, 1, 1, 1]
solution = odeint(z_dot, z_0, time, args=(0,))
plt.plot(time, solution)
plt.title("Uncontrolled nonlinear system")
plt.legend(['x', 'θ', 'x ̇', 'θ ̇'])
plt.show()
```

## Linearized system
$$
\dot{x}=\left[\begin{matrix}0&0&1&0\\0&0&0&1\\0&\frac{g}{5}&0&0\\0&\frac{6g}{5}&0&0\end{matrix}\right]x +
\left[\begin{matrix}0\\0\\\frac{1}{5}\\\frac{1}{5}\\\end{matrix}\right]u
$$
### Controller
$$
u = -Kx
$$
$$
K = \left[\begin{matrix}-12.23&246.1&-25.48&75.48\end{matrix}\right]
$$

```python
A_linear = np.matrix([[0,     0, 1, 0],
                      [0,     0, 0, 1],
                      [0,   g/5, 0, 0],
                      [0, 6*g/5, 0, 0]])
B_linear = np.matrix([[  0],
                      [  0], 
                      [0.2],
                      [0.2]])
# Controller
K = np.matrix([[-12.23, 246.1, -25.48, 75.48]])

controlled_A = A_linear - B_linear@K
def linearized(x, t):
    return np.asarray(controlled_A@x).reshape(-1)
```

## Luenberger Observer
$$
\dot{\hat{x}}=\left[\begin{matrix}-7&0&1&0\\0&-3&0&1\\-12&0&0&0\\0&-2&0&0\\\end{matrix}\right]\hat{x}+\left[\begin{matrix}0\\0\\\begin{matrix}\frac{1}{5}\\\frac{1}{5}\\\end{matrix}\\\end{matrix}\right]u+\left[\begin{matrix}7&0\\0&3\\12&1.96\\0&13.8\\\end{matrix}\right]\left(y-\hat{y}\right)
$$
The $L$ matrix has been determined using the MATLAB command $L = place(A', C', [-1, -2, -3, -4])'$.
The observer's $A$ matrix was calculated as $A_{observer}=A_{linear}-LC$

```python
# Luenberg observer
A = np.matrix([[ -7,  0, 1, 0],
               [  0, -3, 0, 1],
               [-12,  0, 0, 0],
               [  0, -2, 0, 0]])
B = B_linear
L = np.matrix([[1.9,  0.8],
               [0.8,  5.1],
               [1.6,  3.5],
               [2.1, 12.8]])
C = np.matrix([[1, 0, 0, 0],
               [0, 1, 0, 0]])
LC = L@C
def observer(x_hat, t, u, x):
    return (A@x_hat + (B@u).A1 + LC@(x - x_hat)).A1
```

```python
samples = 1000
t0 = 0
tf = 20
time = np.linspace(t0, tf, samples)
x_0 = np.array([1, 1, 1, 1])    # initial conditions
y = [(C@x_0).A1]
xs = [x_0]
x_hats = [x_0]
zs = [x_0]
for i in range(1, samples):
    local_time = np.linspace(time[i-1], time[i])  # for better accuracy on a small time interval
    u = (K@x_0).A1          # calculate control
    noise = 0#np.random.normal()  # Gaussian noise
    noise2 = 0#np.random.normal()  # another Gaussian noise
    x = odeint(linearized, x_0, local_time)   # get the state from the controlled linearized system
    x_0 = x[-1] + noise2    # new starting point is current end point
    xs.append(x_0)    # save the state of the current time snapshot
    y.append((C@x_0).A1) # linearize system output
    x_hat = odeint(observer, x_hats[-1], local_time, args=(u, x_0 + noise))
    x_hats.append(x_hat[-1])
    z = odeint(z_dot, x_0, local_time, args=(u,))   # solve the nonlinear system using the same control
    zs.append(z[-1])
    
y = np.array(y)
xs = np.array(xs)
x_hats = np.array(x_hats)
zs = np.array(zs)
```

```python
plt.figure(figsize=(10, 10))
plt.title("Inverted pendulum on a cart system simulation with Gaussian noise on dynamics")
plt.plot(time, y[:, 0], label="Output x")
plt.plot(time, y[:, 1], label="Output θ")
plt.plot(time, xs[:, 0], label="Linearized controlled x")
plt.plot(time, xs[:, 1], label="Linearized controlled θ")
plt.plot(time, x_hats[:, 0], label="Observer controlled x")
plt.plot(time, x_hats[:, 1], label="Observed controlled θ")
plt.plot(time, zs[:, 0], label="Nonlinear controlled x")
plt.plot(time, zs[:, 0], label="Nonlinear controlled θ")
plt.grid()
plt.legend()
```
