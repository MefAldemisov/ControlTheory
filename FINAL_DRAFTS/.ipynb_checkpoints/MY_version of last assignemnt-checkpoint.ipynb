## Dynamic Parameter estimation

Initial system:

$$m_{11}\ddot{q_1} + m_{12} \ddot{q_2} + \mu_1 \dot{q_1} + k_1 q_1 = u$$
$$m_{21}\ddot{q_1} + m_{22} \ddot{q_2} + \mu_1 \dot{q_2} + k_1 q_2 = 0$$

Representation in the form $Y(\ddot{q}, \dot{q}, q)\Theta = y$:
$$q = \begin{bmatrix} q_1 & q_2\end{bmatrix}^T$$
$$\ddot{q}\begin{bmatrix} m_{11} & m_{12} \\ m_{21} & m_{22} \end{bmatrix} + \dot{q} \begin{bmatrix} \mu_1 & 0 \\ 0 & \mu_1\end{bmatrix} + q \begin{bmatrix}k_1 & 0 \\ 0 & k_1\end{bmatrix} = \begin{bmatrix}u \\ 0\end{bmatrix}$$


$$\begin{bmatrix}\ddot{q_1} \\ \ddot{q_2} \\ \dot{q_1} \\ \dot{q_2} \\ q_1 \\ q_2\end{bmatrix}\begin{bmatrix}m_{11} & m_{12}  & \mu_1 & 0 & k_1 & 0 \\ m_{21} & m_{22}  & 0 & \mu_1 & 0 & k_1\end{bmatrix} = \begin{bmatrix}u \\ 0\end{bmatrix}$$

### Main parameters

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def A(m_11 = 0.01, m_12 = 0.02, m_21 = 0.1, m_22 = 0.21):
    return np.array([[m_11, m_12], [m_21, m_22]])

def B(mu_1 = 0.01):
    return np.identity(2)*mu_1

def C(k_1 = 0.5):
    return np.identity(2)*k_1
```

### SImulation. No control

```python
def p_dot(x, t, u):
    q_dot = np.array([[x[0]], [x[1]]])
    q = np.array([[x[2]], [x[3]]])
    a_inv = np.linalg.inv(A())
    mult = np.array([-B().dot(a_inv), -C().dot(a_inv)])
    q_ddot = mult[0].dot(q_dot) + mult[1].dot(q) + np.array([[u], [0]])
    return [q_ddot[0][0], q_ddot[1][0], q_dot[0][0], q_dot[1][0]]
```

```python
time = np.linspace(0, 10, 1000)
q_0 = [0, 0, 0, 0]
res = odeint(p_dot, q_0, time,  args=(0.1, ))
```

```python
plt.figure(figsize=(8, 8))
plt.title("Simple simultion with constant input")
plt.plot(time, res)
plt.grid()
```

### Simulation. PD control

Given desired tragectory:
$$q_1^* (t) = sin(t)$$
$$q_2^* (t) = cos(t)$$

```python
k_p = np.array([0])
k_d = np.array([130])

def q_desired(t):
    return np.array([[np.sin(t)], [np.cos(t)]])

def q_dot_desired(t):
    return np.array([[np.cos(t)], [-np.sin(t)]])

def pd_control(x, t):
    q_dot = np.array([[x[0]], [x[1]]])
    q = np.array([[x[2]], [x[3]]])
    
    d_q_dot = q_dot_desired(t) - q_dot
    d_q =  q_desired(t) - q
    
    u = k_p.dot(d_q) + k_d.dot(d_q_dot)
    return p_dot(x, t, u)

res = odeint(pd_control, q_0, time)
```

```python
plt.figure(figsize=(8, 8))
plt.title("Simulation with PD control")
plt.plot(time, res)
plt.grid()
```

### Simulation. Dynamic parameter estimation.

Dynamic parameter estimation:
$$\frac{d\hat{\theta}}{dt} = \Gamma M^T (y - M\hat{\theta})$$

Static parameter estimation:
$$\Theta = (M^TM)^{-1}M^Ty$$

```python
# Gamma = np.identity(3)
# theta_initial = np.zeros((2, 6))

res.shape

M = np.array([[0]*6] + [[res[i][0]-res[i-1][0], res[i][1]-res[i-1][1], res[i][0], res[i][1], res[i][2], res[i][3]]  for i in range(1, res.shape[0])])
```

```python
def get_u(x, t):
    q_dot = np.array([[x[0]], [x[1]]])
    q = np.array([[x[2]], [x[3]]])
    
    d_q_dot = q_dot_desired(t) - q_dot
    d_q =  q_desired(t) - q
    
    return k_p.dot(d_q) + k_d.dot(d_q_dot)
 
y = np.array([[get_u(x, t), 0] for x, t in zip(res, time)])
```

```python
np.linalg.inv((M.T).dot(M)).dot(M.T).dot(y)
```
