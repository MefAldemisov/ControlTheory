2) The system should be simulated with some controller... e.g. 
- PD ($u = kp(q^* - q) + kd(\dot{q^*}-\dot{q})$) - do not have desired trajectory
- LQR method or proportional controller -> +

For pole-placement method the A and B matrices should be found

$$\begin{bmatrix}\ddot{x} \\ \dot{x} \\ 0\end{bmatrix} = \begin{bmatrix}\frac{-\mu}{m} & \frac{-k}{m} & \frac{f}{m} \\ 1& 0& 0\\ 0& 0& 0 \end{bmatrix}\begin{bmatrix}\dot{x} \\ x \\ 1\end{bmatrix} + \begin{bmatrix}10\frac{u}{m}\\ 0 \\ 0\end{bmatrix}$$

Does not work, because $M^TM$ is not invertable

$$\begin{bmatrix}\ddot{x} \\ \dot{x} \end{bmatrix} = \begin{bmatrix}\frac{-\mu}{m} & \frac{-k}{m} \\ 1& 0\end{bmatrix}\begin{bmatrix}\dot{x} \\ x \end{bmatrix} + \begin{bmatrix}\frac{10u + f}{m}\\ 0 \end{bmatrix}$$

```python
import scipy.signal as sig
```

```python
m = 10
mu = 1
k = 20
f = 30

A = np.array([[-mu/m, -k/m], [1, 0]])
B = np.array([[1/m, 0]])
```

```python
def get_u(t):
    #some arbitrary control
    return np.sin(t) + 0.1

def simulation(x, t):
    u = 10*get_u(t) + f
    r = A.dot(x) + B[0]*u
    return r

time = np.linspace(0, 10, 100)
res = odeint(simulation, [0, 0], time)

plt.figure(figsize=(8, 8))
plt.title("Sample simulation with constatnt input")
plt.plot(time, res)
plt.grid()
```

3) $\boxed{\hat{\theta} = (M^TM)^{-1}M^Ty}$

```python
y = np.array([(get_u(t)*10 + f) for t in time])
M = np.array([[0, 0, 0]] + [ [res[i][0] - res[i-1][0], res[i][0], res[i][1]] for i in range(1, res.shape[0])])
```

```python
np.linalg.inv(M.T.dot(M)).dot(M.T).dot(y)
```
